```python
from math import exp
 

alpha=1
dataset = np.matrix([[1,3,0],
	[2,2,1],
	[2,1,0],
	[3,1,0],
	[5,4,1]])

# Make a prediction with coefficients
def predict(dataset):
    sum=0 
    coef = np.array([0,0,0])
    for data in dataset:
        sum += (1/(1 + exp(-(coef[0] + coef[1]*data.item(0) + coef[2]*data.item(1))))) - data.item(2)
    print(sum)
    coef[0] -= alpha*sum
    sum=np.array([0.,0.])
    for data in dataset:
        sum += np.array([data.item(0),data.item(1)]).dot((1/(1 + exp(-(coef[0] + coef[1]*data.item(0) + coef[2]*data.item(1)))) - data[2])))
    sum=-1*sum
    coef[1] -= alpha*sum[0]
    coef[2] -= alpha*sum[1]
    return coef
    
print(predict(dataset))
```
