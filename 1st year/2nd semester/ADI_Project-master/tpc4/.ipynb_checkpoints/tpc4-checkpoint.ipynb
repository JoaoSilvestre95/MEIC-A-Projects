```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

plt1 = plt.figure(1)




plt.scatter(1,3, marker="o", c="red")
plt.scatter(2,2, marker="x", c="blue")
plt.scatter(2,1, marker="o", c="red")
plt.scatter(3,1, marker="o", c="red")
plt.scatter(5,4, marker="x", c="blue")
plt.title("Plot")   
plt.xlabel("Attribute 1")
plt.ylabel("Attribute 2")

plt1.show()
```

```python
from math import exp
 
coef = np.array([0,0,0])
alpha=1
dataset = [[1,3,0],
	[2,2,1],
	[2,1,0],
	[3,1,0],
	[5,4,1]]

# Make a prediction with coefficients
def predict(dataset, coef):
    sum=0    
    for data in dataset:
        sum += (1/(1 + exp(-(coef[0] + coef[1]*data[0] + coef[2]*data[1])))) - data[2]
    sum=-1*sum    
    coef[0] -= alpha*sum
    print(coef)
    sum=np.array([0.,0.])
    for data in dataset:
        sum += np.array([0,0])*(((1/(1 + exp(-(coef[0] + coef[1]*data[0] + coef[2]*data[1])))) - data[2]))
    sum=-1*sum
    coef[1] -= alpha*sum[0]
    coef[2] -= alpha*sum[1]
    return coef
    
print(predict(dataset, coef))
```
